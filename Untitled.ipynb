```r
library(digest)
library(gridExtra)
library(repr)
library(tidyverse)
library(readxl)
options(repr.matrix.max.rows = 8)
```

```r
#Load the data
data <- read_csv("Blue Chip Credit Dataset.csv")
data
```

#### Cleaning the data:

```r
#Check if the data have any missing value
sum(is.na(data))
# The data does not have any missing value, then checking for the duplicated value in the data set
sum(duplicated(data))
clean_data <- unique(data)
clean_data
```

#### Visualize for the data:

```r
#Based on the Attrition Flag
ggplot(clean_data, aes(x = Attrition_Flag, fill = Attrition_Flag)) +
  geom_bar() + 
labs(x = "Attrition Flag", y = "Count") 


attrition_rate_total <- clean_data %>%
    group_by(Attrition_Flag) %>%
    summarise(n = n(), 
           percent = n / nrow(clean_data) * 100)

attrition_rate_total
```

```r
#Based on the Card Category
ggplot(clean_data, aes(x = Card_Category, fill = Card_Category)) +
       geom_bar() + 
       labs(x = "Card Category", y = "Count")  + 
       scale_fill_manual(values=c("#0000FF","#FFD700","#000000","#C0C0C0"))

attrition_rate_total_card_type <- clean_data %>%
    group_by(Card_Category) %>%
    summarise(n = n(), 
           percent = n / nrow(clean_data) * 100)

attrition_rate_total_card_type
```

```r
#Based on Card Category and Attrition Flag
options(repr.plot.width = 8, repr.plot.height = 6)
ggplot(clean_data, aes(x = Card_Category, fill = Attrition_Flag)) +
    geom_bar(position = position_fill(reverse=TRUE)) 
```

```r
#Based on Gender:
ggplot(clean_data, aes(x = Gender, fill = Gender)) +
  geom_bar() + 
labs(x = "Gender", y = "Count") 
```

```r
#Based on Education Level
ggplot(clean_data, aes(x = Education_Level, fill = Education_Level)) +
  geom_bar() + 
labs(x = "Education Level", y = "Count") 
```

```r
#Based on Income
ggplot(clean_data, aes(x = Income_Category, fill = Income_Category)) +
  geom_bar() + 
labs(x = "Income Category", y = "Count") 

attrition_rate_income_category <- clean_data %>%
    group_by(Income_Category) %>%
    summarise(n = n(), 
           percent = n / nrow(clean_data) * 100)

attrition_rate_income_category
```

```r
summary(clean_data$Customer_Age)
boxplot(clean_data$Customer_Age)
data_age <- ggplot(clean_data, aes(x = Customer_Age)) +
  geom_bar() + 
labs(x = "Customer Age", y = "Count") 
```

```r
#Divide customer data into age group
clean_data$age_group[clean_data$Customer_Age <= 25] <- "18 - 25"
clean_data$age_group[(clean_data$Customer_Age >= 26) & (clean_data$Customer_Age <= 33)] <- "26 - 33"
clean_data$age_group[(clean_data$Customer_Age >= 34) & (clean_data$Customer_Age <= 41)] <- "34 - 41"
clean_data$age_group[(clean_data$Customer_Age >= 42) & (clean_data$Customer_Age <= 49)] <- "42 - 49"
clean_data$age_group[(clean_data$Customer_Age >= 50) & (clean_data$Customer_Age <= 57)] <- "50 - 57"
clean_data$age_group[(clean_data$Customer_Age >= 58) & (clean_data$Customer_Age <= 65)] <- "58 - 65"
clean_data$age_group[(clean_data$Customer_Age >= 66) & (clean_data$Customer_Age <= 73)] <- "66 - 73"

ggplot(clean_data, aes(x = age_group, fill= Attrition_Flag)) + 
geom_bar(position = position_fill(reverse = TRUE))
```

```r
#Relation between Months on Book and Customer Age
clean_data %>%
  ggplot(aes(x = Customer_Age, y = Months_on_book, color = Attrition_Flag)) +
  labs(x = "Customer Age", y = "Months on book") +
  geom_point()
```

```r
#Relation between Months Inactive

ggplot(clean_data, aes(x = Months_Inactive_12_mon, fill = Attrition_Flag)) +
    geom_bar(position = position_fill(reverse=TRUE)) 
```

#### Young people (from 18 - 25): ALL OF THEM DROP!!!

```r
young_people <- filter(attrited_data, age_group == "18 - 25")
young_people
```

**Conclusion**

We can see that all of the young customer that have had stopped using company's card.

There are many reason of why all of them have left:
1. They having the same education level  which is college (maybe they are still go to school) and have low income (Less than $40K) => their demand to use credit card is not really that necessary.
2. Most of them using Blue card and Silver card,, so the annual fee could be too high for people who do not use credit card that much.
3. The benefit that they could earn from those type of cards is not worth compares to the annual fee that they have to pay for each year. For Blue card, the Cashback benefit has high rate but they only have Cashback for first 3 months, which can attracted young people at first but then after 3 months, no longer. This cashback value is too short. For Silver card, they only earn 1% Cashback for gas station payment (not many students often drives to school) not really suitable for young people needs.

**Suggest solution:**

Make a new type of card for student: Example Student Credit Card with no annual fee, maybe have Cash back (for longer time for example first 12 months with 5% Cashback or Cashback all the time, lower rate (2% for all type/ or 3% for grocery and student supplies and 1% for other types). Have reward points program to purchase other things.
