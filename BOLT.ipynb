```r
library(tidyverse)
library(repr)
library(readxl)
options(repr.matrix.max.rows = 8)
```

```r
data <- read_csv("Blue Chip Credit Dataset.csv")
#Remove duplicated
data
data_set <- distinct(data, CLIENTNUM, Card_Category, .keep_all= TRUE)
data_set
```

```r
attrited_customer_data <- filter(data_set, Attrition_Flag == "Attrited Customer")
attrited_customer_data
num_attrited <- nrow(attrited_customer_data)
```

## 1. Based on type of Card:

```r
#Find the portion of each type of card in Attrited Customer:
blue_card_attrited_customer_data <- filter(attrited_customer_data, Card_Category == "Blue")
blue_card_attrited_customer_data
num_blue_attrited <- nrow(blue_card_attrited_customer_data)
```

```r
silver_card_attrited_customer_data <- filter(attrited_customer_data, Card_Category == "Silver")
silver_card_attrited_customer_data
num_silver_attrited <- nrow(silver_card_attrited_customer_data)
```

```r
gold_card_attrited_customer_data <- filter(attrited_customer_data, Card_Category == "Gold")
gold_card_attrited_customer_data
num_gold_attrited <- nrow(gold_card_attrited_customer_data)
```

```r
platinum_card_attrited_customer_data <- filter(attrited_customer_data, Card_Category == "Platinum")
platinum_card_attrited_customer_data
num_platinum_attrited <- nrow(platinum_card_attrited_customer_data)
```

```r
slices <- c(num_blue_attrited, num_silver_attrited,num_gold_attrited, num_platinum_attrited)
lbls <- c("Blue", "Silver", "Gold", "Plantinum")
pie(slices, labels = lbls, main="Pie Chart of Card")
```


=> Most of the Attrited Customer having Blue Card => Explore more data of Blue Card (maybe with Silver Card)
